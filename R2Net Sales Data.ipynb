```python
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
import pandas as pd
```

```python
plt.style.use('fivethirtyeight')
```

```python
sales_data= pd.read_csv('sales_data_sample .csv')
```

```python
sales_data.head(10)
```

```python
sales_data.dtypes
```

```python
sales_data['SALES'] = sales_data['SALES'].astype(int)
sales_data['PRICEEACH'] = sales_data['PRICEEACH'].astype(int)
```

```python
sales_data['ORDERDATE'] = pd.to_datetime(sales_data['ORDERDATE'])
```

```python
sales_data['STATUS'] = sales_data['STATUS'].replace('Shppd', 'Shipped')
sales_data['STATUS'] = sales_data['STATUS'].replace('Canceled', 'Cancelled')
```

```python
sales_data['Month'], sales_data['Year'] = sales_data['Month-Year'].str.split("-", 1).str
```

```python
sales_data.columns
```

```python
sales_data = sales_data[['ORDERNUMBER', 'QUANTITYORDERED', 'PRICEEACH', 'ORDERLINENUMBER',
       'SALES', 'ORDERDATE', 'STATUS', 'QTR_ID', 'Month-Year', 'Month', 'Year','PRODUCTLINE',
       'MSRP', 'PRODUCTCODE', 'CUSTOMERNAME', 'PHONE', 'ADDRESSLINE1',
       'ADDRESSLINE2', 'CITY', 'STATE', 'POSTALCODE', 'COUNTRY', 'TERRITORY',
       'CONTACTLASTNAME', 'CONTACTFIRSTNAME', 'DEALSIZE']]
```

```python
sales_data.head(10)
```

```python
avg_quantity = sales_data['QUANTITYORDERED'].mean()
avg_quantity
```

```python
most_popular = sales_data.groupby('PRODUCTCODE')[['QUANTITYORDERED']].sum()
most_popular.sort_values(by = 'QUANTITYORDERED',ascending=False)
```

```python
status = sales_data.groupby('STATUS')[['ORDERNUMBER']].count()
status
```

```python
status['Average'] = ((status['ORDERNUMBER'] / 2823) * 100).round(2) 
status['Average'] = status['Average'].astype(str) + '%'
status
```

```python
deal_sales = sales_data.groupby('DEALSIZE')[['SALES']].mean().round(2)
deal_sales
```

```python
deal_quantity = sales_data.groupby('DEALSIZE')[['QUANTITYORDERED']].sum()
deal_quantity
```

```python
top_five = sales_data.groupby(['DEALSIZE','CUSTOMERNAME'])[['CUSTOMERNAME']].count().rename(columns = {'CUSTOMERNAME':'COUNT'})
top_five
```

```python
top_five = top_five.reset_index()
top_five
```

```python
fivefive = top_five.sort_values(['DEALSIZE','COUNT'],ascending=False).groupby('DEALSIZE').head(5)
fivefive
```

```python
quarter = sales_data.groupby('QTR_ID')[['QUANTITYORDERED','SALES']].sum()
quarter
```

```python
region = sales_data.groupby('COUNTRY')[['QUANTITYORDERED','SALES']].sum()
region.sort_values(by='SALES',ascending=False)
```

```python
sales_data['PRICEEACH'] = '$' + sales_data['PRICEEACH'].astype(str)
sales_data['SALES'] = '$' + sales_data['SALES'].astype(str)
```

```python
sales_data
```

```python
sales_data.to_csv('cleaned_data.csv')
```
